## Visualizations graphs

### 1. Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x='Legislative District', y='Electric Range', data=electric_df)
plt.xlim(left=0, right=100)  
plt.ylim(bottom=0, top=500)  
plt.title('Legislative District Vs Electric Range')
plt.xlabel('Legislative District')
plt.ylabel('Electric Range')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 2. Heatmap

```python

numeric_df = electric_df.select_dtypes(include = 'number')
plt.figure(figsize = (12,7))
sns.heatmap(numeric_df.corr(),annot = True , vmin = -1, vmax = 12)
plt.show()
```

### 3. Bar Plot

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Electric Vehicle Type', y='Base MSRP', data=electric_df)
plt.title('Base MSRP of Top 5 Electric Vehicles by Type')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Base MSRP')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4. Pie Chart

```python

make_counts = electric_df['Make'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Distribution of Car Makes')
plt.show()
```

### 5. Histogram

```python
plt.figure(figsize = (10,6))
plt.hist(electric_df['Electric Range'].dropna(),bins = 10,color = 'blue',edgecolor = 'black')
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range (km)')
plt.ylabel('frequency')
plt.tight_layout()
plt.show()
```
